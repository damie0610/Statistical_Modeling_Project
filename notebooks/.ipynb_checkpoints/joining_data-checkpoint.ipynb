# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
# Import the necessary libraries and packages
import pandas as pd
import numpy as np
import sqlite3
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline     
sns.set(color_codes=True)
```

```python
# Import CSV files into separate DataFrames
df1 = pd.read_csv('C:/Users/HP/Music/LHLDataCourse/Python/project_data/ph_bikes.csv')
df2 = pd.read_csv('C:/Users/HP/Music/LHLDataCourse/Python/project_data/yelp_fit.csv')
df3 = pd.read_csv('C:/Users/HP/Music/LHLDataCourse/Python/project_data/yelp_uni.csv')
df4 = pd.read_csv('C:/Users/HP/Music/LHLDataCourse/Python/project_data/fsq_fit.csv')
df5 = pd.read_csv('C:/Users/HP/Music/LHLDataCourse/Python/project_data/fsq_uni.csv')
```

```python
# Specify the path to the SQLite database file
db_path = "C:/Users/HP/Music/LHLDataCourse/Python/project_data/bike_master.db"

# Create a connection to SQLite database
conn = sqlite3.connect(db_path)
```

```python
# Write DataFrames to SQLite tables
df1.to_sql('table1', conn, index=False, if_exists='replace')
df2.to_sql('table2', conn, index=False, if_exists='replace')
df3.to_sql('table3', conn, index=False, if_exists='replace')
df4.to_sql('table4', conn, index=False, if_exists='replace')
df5.to_sql('table5', conn, index=False, if_exists='replace')
```

```python
# SQL query using joins to merge the tables
sql_query = """
SELECT *
FROM table1
INNER JOIN table2 ON table1.ID = table2.ID
INNER JOIN table3 ON table1.ID = table3.ID
INNER JOIN table4 ON table1.ID = table4.ID
INNER JOIN table5 ON table1.ID = table5.ID
"""
master_df = pd.read_sql_query(sql_query, conn)
master_df = master_df.drop("ID", axis=1)

conn.close()
```

```python
# Display the merged DataFrame
pd.set_option('display.max_columns', None)
master_df.head()
```

```python
csv_file_path = 'master_df.csv'

# Save the DataFrame to a CSV file
master_df.to_csv(csv_file_path, index=False)
```

### Adding Columns for Analysis

##### I created 4 columns: Total number of bike slots (*total_slots*), Percent usage of bikes (*pct_usage*), Category for High/Low Usage and a binary representation for the category (*usage_cat* & *bin_usage*)

```python
master_df["total_slots"] = master_df["empty_slots"] + master_df["free_bikes"]
master_df["pct_usage"] = (master_df["empty_slots"]/master_df["total_slots"])*100
master_df.info()
```

```python
# Check the mean percentage usage
master_df['pct_usage'].describe()
```

```python
# Creating categories for usage based on pct_usage
# Since the mean is 64%, my assumption is that usage higher than the mean is High and otherwise is low.

threshold = 64  # threshold for "high usage"

# new column 'usage_cat' based on the threshold
master_df['usage_cat'] = master_df['pct_usage'].apply(lambda x: 'High' if x > threshold else 'Low')

# new column 'bin_usage' with 1 for "high usage" and 0 for "low usage"
master_df['bin_usage'] = master_df['pct_usage'].apply(lambda x: 1 if x > threshold else 0)

# Display the DataFrame with the new columns
master_df.head()
```

```python
# Correlations of numeric variables with pct_usage (dependent variable)
pairplot = sns.pairplot(
    master_df, 
    x_vars=["pct_usage", "yelp_center_dist", "yelp_college_dist", "fsq_center_dist", "fsq_college_dist"],
    y_vars=["pct_usage", "yelp_center_dist", "yelp_college_dist", "fsq_center_dist", "fsq_college_dist"],
    diag_kind='hist',
    hue='usage_cat')

plt.show()
```

```python
# Checking the normality of the dependent variable
pct_hist = sns.histplot(master_df['pct_usage'])
plt.show()
```

```python
# Save the histogram as an image
pct_hist.figure.savefig('pct_hist.png')
```

```python
from scipy import stats
stat, p = stats.shapiro(master_df['pct_usage'])
print('%0.30f' % p)
```

Since p<0.05, we reject the null hypothesis that the data is normally distributed.

```python
# Correlation check
master_df.corr(numeric_only=True)
```

```python
# Checking the significance of the correlations.

import scipy

stat, p1 = scipy.stats.pearsonr(master_df['pct_usage'], master_df['yelp_center_dist'])
stat, p2 = scipy.stats.pearsonr(master_df['pct_usage'], master_df['yelp_college_dist'])
stat, p3 = scipy.stats.pearsonr(master_df['pct_usage'], master_df['fsq_center_dist'])
stat, p4 = scipy.stats.pearsonr(master_df['pct_usage'], master_df['fsq_college_dist'])

print(p1, p2, p3, p4)
```

Significant correlations observed with the fitness centers, not the colleges.
